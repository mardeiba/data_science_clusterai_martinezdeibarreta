# Modelos de regresi√≥n

```python
#Nuestra hipotesis es que con ciertas features, es posible predecir la produccion de petroleo crudo
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.svm import LinearSVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Debido a la naturaleza de nuestros datos, no necesitamos crear dummies
```

```python
#Creamos la variable dependiente Y
y = np.array(datafilmensual1[["prod_petcru"]])
```

```python
#Creamos la variable independiente X
x = datafilmensual1.drop(['prod_petcru','precint_gasnat','precint_petcru','alcance_nombre','alcance_id','distribution_identifier','indice_tiempo'], axis=1)
```

```python
#Dividimos el dataset entre Train y Test
#Decidimos probar nuestra hipotesis comparando la prediccion que haga nuestro modelo del bottom 20% de los datos
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=22)
```

```python
#Observamos las features que dejamos para que el modelo aprenda y prediga
xtrain.head(5)
```

```python
#Auto scaling train-set 
scaler = preprocessing.StandardScaler().fit(xtrain)
scaler
```

```python
#Auto scalo mis muestras de train utilizando el scaler fiteado con el xtrain
xtrain_scal = scaler.transform(xtrain)  
```

```python
#Auto scalo mis muestras de train utilizando el scaler fiteado con el xtrain
xtest_scal = scaler.transform(xtest)  
```

### Regresi√≥n lineal simple

```python
#Primero hacemos una regresi√≥n lineal simple
lr = LinearRegression()
```

```python
#Ajustamos el modelo utilizando el set de train, tanto las features X(escaladas) como las samples Y
lr.fit(xtrain_scal, ytrain)
```

```python
#Obtengo las predicciones que realiza mi modelo, sin mostrarle las labels (Ytest) y las guardamos en ypred
ypred = lr.predict(xtest_scal)
```

```python
np.shape(ypred)
```

```python
#calculamos el error RMSE de mi modelo con las muestras de train -> cuando le doy importancia a grandes errores
np.sqrt(mean_squared_error(ytest, ypred))
```

```python
#calculamos MSE
mean_squared_error(ytest, ypred)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
#Calculamos MAE
mean_absolute_error(ytest, ypred)
```

   ### Creo un modelo KNN Regresion

```python
#Como segundo modelo, probamos un KNN
prod = KNeighborsRegressor(weights = "distance")
```

```python
#Creo un array para probar por gridsearch con los valores de k
parametrosk= np.arange(20,31,5)
parametrosk
```

```python
#Indico los valores del  parametro que quiero probar
parameters_knn = [{'n_neighbors': parametrosk}]
```

```python
#Folds para el training del CV
n_folds = 6
```

```python
#Creo un modelo que contiene: el regresor, los parametros a probar y los folds para el CV
regresor_prod = GridSearchCV(prod, parameters_knn, refit = True, cv=n_folds, verbose=True)
```

```python
#Entreno modelo regresor con mi set de entrenamiento 
regresor_prod.fit(xtrain_scal, ytrain)
```

```python
#muestro el mejor hiperparametro segun el cross validation y grid search
regresor_prod.best_params_
```

```python
#calcula el R2
regresor_prod.best_score_
```

```python
#Predigo Y (ypred) usando solo xtest con mi modelo
ypred2 = regresor_prod.predict(xtest_scal)
ypred2
```

```python
#Miro los datos reales (ytest) para poder compararlos con la prediccion obtenida
ypred2 = regresor_prod.predict(xtest_scal)
ypred2
```

```python
fig, ax = plt.subplots(figsize=(30, 10))
plt.plot(ytest, c="green")
plt.plot(ypred2, c="red")
plt.ylim(0, 1500000)
plt.ylabel('Producci√≥n')
plt.show()
```

```python
#Calculo el RMSE
np.sqrt(mean_squared_error(ytest, ypred2))
```

```python
#Calculo el MSE
mean_squared_error(ytest, ypred2)
```

```python
#Calculo MAE
mean_absolute_error(ytest, ypred2)
```

### Hacemos la regresion con el modelo SVR

```python
#Finalmente probamos con un modelo SVR
svreg = SVR()
```

```python
#Cantidad de folds para el CV
n_folds = 6
```

```python
#Hiper-parametros a probar en GS (pruebo gridsearch por kernels separados)
parameters_svr_lin = [{'kernel':['linear'], 'C': [1, 100]}]

parameters_svr_rbf = [{'kernel':['rbf'] , 'C': [1,100],'gamma': [0.1,0.5] }]
```

```python
regresor_svr = GridSearchCV(svreg, parameters_svr_lin, cv=n_folds, verbose=True)
```

```python
#Entreno modelo regresor con mi set de entrenamiento 
regresor_svr.fit(xtrain_scal, ytrain.ravel())
```

```python
#Mejores parametros luego del CV
regresor_svr.best_params_
```

```python
#Calculo R2
regresor_svr.best_score_
```
